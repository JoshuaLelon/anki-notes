```python
df2['text'] = df[0].str[10:]
df2.head()
```

```python
def swap_with_numbers(x):
    if x == "__label__2":
        return 1
    elif x == "__label__1":
        return 0
    else:
        return None
    
df2['label'] = df2['label'].apply(swap_with_numbers)
df2.head()
```

```python
df = pd.read_csv("datasets/twitter-airline-sentiment.csv")
df = df[ ['airline_sentiment', 'text'] ]
df.head()
```

```python
def convert_to_numbers(x):
    if x == 'positive':
        return 1
    elif x == 'negative':
        return 0
    else:
        return -1

df['airline_sentiment'] = df['airline_sentiment'].apply(convert_to_numbers)
df.head()
```

```python
print(len(df['text']))
df = df[df.airline_sentiment != -1]
print(len(df['text']))
df.head()
```

```python
df.rename(columns={'airline_sentiment':'label'}, inplace=True)
df.head()
```

```python
df.to_csv("datasets/twitter-airline-sentiment.csv", index=False)
```

# Sampling: Creating <100mb train and val sets

```python
import pandas as pd

dataset_dirs = ['twitter.csv', 'amazon.csv', 'imdb.csv']
for dataset in dataset_dirs:
    df = pd.read_csv('datasets/' + dataset)
    train = df.sample(frac=0.8, random_state=200)
    val = df.drop(train.index)
    train.to_csv('datasets/train_' + dataset, index=False)
    val.to_csv('datasets/val_' + dataset, index=False)
    
```
