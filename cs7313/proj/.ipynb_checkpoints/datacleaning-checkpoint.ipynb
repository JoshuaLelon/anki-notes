```python
# read in yelp JSON file, convert to csv
import json
import csv
import pandas as pd

amazon_txt = "datasets/amazon-reviews.txt"
amazon_csv_path = "datasets/amazon-reviews.csv"


df = pd.read_csv(amazon_txt, sep='sdafasdfoaiwf', header=None)
df.head()
# df.to_csv(amazon_csv_path)
```

```python
df2 = pd.DataFrame()
df2['label'] = df[0].str[:10]
```

```python
df2.head()
```

```python
df2['text'] = df[0].str[10:]
df2.head()
```

```python
def swap_with_numbers(x):
    if x == "__label__2":
        return 1
    elif x == "__label__1":
        return 0
    else:
        return None
    
df2['label'] = df2['label'].apply(swap_with_numbers)
df2.head()
```

```python
df2.to_csv(amazon_csv_path, index=None)
```

```python
df = pd.read_csv("datasets/twitter-airline-sentiment.csv")
df = df[ ['airline_sentiment', 'text'] ]
df.head()
```

```python
def convert_to_numbers(x):
    if x == 'positive':
        return 1
    elif x == 'negative':
        return 0
    else:
        return -1

df['airline_sentiment'] = df['airline_sentiment'].apply(convert_to_numbers)
df.head()
```

```python
print(len(df['text']))
df = df[df.airline_sentiment != -1]
print(len(df['text']))
df.head()
```

```python
df.rename(columns={'airline_sentiment':'label'}, inplace=True)
df.head()
```

```python
df.to_csv("datasets/twitter-airline-sentiment.csv", index=False)
```
